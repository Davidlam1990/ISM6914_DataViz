```python
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
def singlename(fn, ln):
    return fn + " " + ln
df['fullname'] = singlename(df['fname'],df['lname'])
```
