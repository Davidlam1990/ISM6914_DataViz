```python
#Applying Functions to Groups, Bins, and Columns
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
# Create the bin dividers
bins = [0, 60, 70, 80, 90, 100]
# Create names for the four groups
group_names = ['F', 'D', 'C', 'B', 'A']
df['letterGrades'] = pd.cut(df['grade'],bins, labels=group_names)
df.head()
```

```python
df.groupby('letterGrades')['hours'].mean()
```

```python
# Applying the integer function to the grade column
df['grade'] = df['grade'] = df['grade'].apply(lambda x: int(x))
df.head()
```

```python
gender_preScore = df['grade'].groupby(df['gender'])
gender_preScore.mean()
```

```python
#Your Turn
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
# Create the bin dividers
bins = [0,70,100]
# Create names for the four groups
group_names = ['Fail', 'Pass']
df['status'] = pd.cut(df['grade'],bins, labels=group_names)
df.tail()
```

```python
# Create new passign_females df for 'female' and 'pass'

pass_females = df.loc[(df['gender']=='female') & (df['status']=='Pass')]
pass_females.head()
```

```python
pd.value_counts(pass_females['gender'])
```

```python
pd.value_counts(pass_females['status'])
```
