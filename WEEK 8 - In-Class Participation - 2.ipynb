```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
Location = "datasets/covid.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df2 = df[df['Province_State'] == "Florida"]
df2.head()
```

```python
# Ten largest values in column Weight 
top10deathFL_df = df2.nlargest(10, ['Deaths']) 
top10deathFL_df
```

```python
mydf = pd.DataFrame(data = top10deathFL_df,
columns=['Combined_Key', 'Deaths'])
%matplotlib inline
mydf2 = mydf.set_index(mydf['Combined_Key'])
mydf2.plot(kind="bar")
plt.title('Top 10 deaths in FL by City on 4/1/2020')
```
